```python
df = spark.read.table("stockprices.01_bronze.daily_prices")
```

```python
from pyspark.sql.functions import to_date, col, round
df = df.drop("trade_count", "vwap")
df = df.withColumn("timestamp", to_date("timestamp").cast("date"))\
    .withColumnRenamed("symbol", "ticker")\
    .withColumn("open", round(col("open"), 2))\
    .withColumn("high", round(col("high"), 2))\
    .withColumn("low", round(col("low"), 2))\
    .withColumn("close", round(col("close"), 2))\
    .withColumn("volume", col("volume").cast("int"))\
    .withColumn("daily_return", round((col("close") - col("open")) / col("open"), 4))\
    .withColumn("price_range", round(col("high") - col("low"), 2))

```

```python
df.write.mode("overwrite").saveAsTable("stockprices.02_silver.daily_prices_cleaned")
```
